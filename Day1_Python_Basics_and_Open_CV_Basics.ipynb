**Imports**

```python
import cv2
import matplotlib.pyplot as plt
```

**For reading image we need to copy- paste the path of the image.**

```python
image = cv2.imread('/content/face.jpg', 0)
```

```python
print(image[:200])
```

**Displaying Properties of Images**

```python
print(image.shape)
```

```python
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

plt.imshow(image[:200])
plt.title("Face")
plt.show()
```

**For accessing the camera**

```python
cap = cv2.VideoCapture(0)

while True:
	flag, frame = cap.read()
	if not flag:
		print('Could not access the camera')
		break
	cv2.imshow('frame', frame)
	k = cv2.waitkey(1)
	if k == ord('q'):
		break


cap.release()
cv2.destroyAllWindows()
```
